## Preliminaries

We know the definition of determinant

\begin{equation}
  \det A = \sum_{\mu_1\cdots \mu_n} \epsilon_{\mu_1 \cdots \mu_n} a_{1\mu_1} \cdots a_{n\mu_n}
\end{equation}

By the definition of levi-civita symbol (at least two indices are same, then 0),

We can re-write definition as below :

\begin{equation}
  \sum_{\mu \in S_n} \epsilon_{\mu_1 \cdots \mu_n} a_{1\mu_1} \cdots a_{n\mu_n}
\end{equation}

* $\mu$ is permutation of $(1,2,\cdots,n)$ & $S_n$ is n-dimensional permutation group.
* $\mu \, = \, (\mu_1,\cdots, \mu_n)$ where $\mu(i) = \mu_i$
* $\mu$ can be represented by transpositions
> \begin{equation}
   \tau(\mu_i, \mu_j)(k) = \left(
   \begin{array}{ll}
    \mu_j & \mbox{($k = \mu_j$)} \\
    \mu_i & \mbox{($k = \mu_i$)} \\
    k & \mbox{(others)}
   \end{array}
   \right.
  \end{equation}
  
For example, consider $(3\,1\,2\,5\,4)$
> * Permutation
 \begin{align}
   &(1\, 2\, 3\, 4\, 5) \quad \mbox{change 1$\leftrightarrow$3 }\\
  \Rightarrow\, &(3\, 2\, 1\, 4\, 5) \quad \mbox{change 2$\leftrightarrow$1 }\\
  \Rightarrow\, &(3\, 1\, 2\, 4\, 5) \quad \mbox{change 4$\leftrightarrow$5 }\\
  \Rightarrow\,&(3\, 1\, 2\, 5\, 4)
 \end{align}
   
> * Transpostion  
> $\Rightarrow \, \tau(4,5) \tau(1,2) \tau(1,3)$

## Permutation Theorem
**<U>Thm</U>** *If $\mu' = (\mu_2, \cdots, \mu_n)$ is permutation of $(1,\cdots,\mu_1 - 1, \mu_1 + 1, \mu_n)\,$ then for $\,\mu\,=\,(\mu_1, \cdots , \mu_n)$, $\,\epsilon_{\mu_1 \cdots \mu_n} = (-1)^{\mu_1 + 1} \epsilon_{\mu_2 \cdots \mu_n}$*
> **<U>Pf</U>**  
Consider $(\mu_1,\, 1,\,2,\,\cdots,\,\mu_1 - 1,\, \,\mu_1 + 1,\,n)$.  
Then we can represent this by $\tau(1,2)\tau(1,3)\cdots \tau(1,\mu_1 -1)\tau(1,\mu_1)$  

>And let $\mu' =  (\mu_2,\,\cdots,\, \mu_n)\,$ which is permutation of $(1,\,\cdots,\,\mu_1-1,\,\mu_1+1,\,\cdots,\, n)$. Then we can represent this by
$\mu'=\tau(\sim)\tau(\sim)\cdots\tau(\sim)\,$  ($l$ transpositions)

> There is no transposition of $\mu_1$ in $\mu'$, we can write $\mu$ as below:
\begin{equation}
  \mu = \mu' \cdot \tau(1,2)\tau(1,3)\cdots\tau(1,\mu_1)
\end{equation}

> \begin{equation}
   \therefore \, \epsilon_{\mu_1 \cdots \mu_n} = \epsilon_{\mu_2 \cdots \mu_n} \cdot (-1)^{\mu_1 +1}
  \end{equation}
  
**<U>Cor</U>** *If $\mu' = (\mu_1,\,\cdots,\,\mu_{j-1},\, \mu_{j+1},\, \cdots,\, \mu_n)\,$ is permutation of $(1,\,\cdots,\,n)\,$ then $\epsilon_{\mu_1 \cdots \mu_n} = (-1)^{\mu_j + j} \epsilon_{\mu_1 \cdots \mu_{j-1} \mu_{j+1} \cdots \mu_n}$*

**<U>Thm</U>** **Laplace Expansion**  
\begin{equation}
  \det A = \sum_i a_{ij} C_{ij} = \sum_i a_{ji} C_{ji}
\end{equation}
> **<U>Pf</U>**
\begin{align}
  \det A &= \sum_{\mu} \epsilon_{\mu_1\cdots\mu_n} a_{1\mu_1}\cdots a_{j\mu_j}\cdots a_{n\mu_n} \\
  & = \sum_{\mu_j}\sum_{\mu_j \notin \mu'} \epsilon{\mu_1\cdots\mu_n}a_{j\mu_j}a_{1\mu_1}\cdots a_{n\mu_n} \\
  & = \sum_{\mu_j} (-1)^{\mu_j + j} a_{j\mu_j} \sum_{\mu'} \epsilon_{\mu_1\cdots\mu_{j-1}\mu_{j+1}\mu_n} a_{1\mu_1}\cdots a_{j-1\mu_{j-1}}a_{j+1\mu_{j+1}}\cdots a_{n\mu_n} \\
  & = \sum_{i} (-1)^{i+j} a_{ji} M_{ji} \\
  & = \sum_{i} a_{ji} C_{ji} 
\end{align}
